## Opening file using request Library and assigning it to a varribale

```python
#opening the file using Library and then assigning it to a varriable.

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

r = requests.get(url)
content =r.content
```

## Converting HTML format to BeautifulSoup Tree Format

```python

soup = BeautifulSoup(content,'html.parser') 
```

## Finding Relevent data and Assigning them to Lists

```python
A = []
B = []
C = []


for row in soup.findAll('tr'):
    cells = row.findAll('td')  
    
    
    if len(cells) == 3:
        if cells[1].find(text=True).rstrip()=='Not assigned':
            continue
        else :
            A.append(cells[0].find(text=True).rstrip())
            B.append(cells[1].find(text=True).rstrip())
            C.append(cells[2].find(text=True).rstrip())

```

## Assigning Lists into Columns

```python

df = pd.DataFrame(A,columns = ['postal_code'])
df['Brough'] = B
df['Neighborhood'] = C

df
```

## Replacing '/' value with ',' in Neighborhoods Column

```python
df['Neighborhood'] = df['Neighborhood'].str.replace('/' ,',')
df
```
